## Top Performer Analysis

```python
top_batsmen=df.groupby('batsman')["batsman_runs"].sum().reset_index()
top_batsmen=top_batsmen.sort_values(by='batsman_runs',ascending=False)
top_batsmen_10=top_batsmen.head(10)

import plotly.express as px
fig = px.bar(
top_batsmen_10,
x="batsman",
y="batsman_runs",
title="Top 10 Batsman ",
labels={"batsman_runs": "Total Runs", "batsman": "Batsman"},
text="batsman_runs"    
)

fig.show()
```

### Bowler vs Batsman Dismissal 

```python
wicket=df[df['dismissal_kind'].notna()]
bowl_batsmen=wicket.groupby(['bowler','batsman']).size().reset_index(name='dismissals')
top_matchups = bowl_batsmen.sort_values(by='dismissals', ascending=False).head(10)

fig = px.imshow(
    top_matchups.pivot(index='bowler', columns='batsman', values='dismissals'),
    text_auto=True,
    color_continuous_scale="Reds",
    labels=dict(x="Batsman", y="Bowler", color="Dismissals"),
    title="Top Bowler vs Batsman Dismissals"
)
fig.update_layout(width=900, height=600)
fig.show()
```

 ## *Insight*: 
- The heatmap highlights SP Narineâ€™s strong record against V Kohli (136 dismissals) and RA Jadejaâ€™s success against  SK Raina (146 dismissals), showing targeted effectiveness.  
 ## *Action*:
- Coaches should deploy Narine against Kohli and Jadeja against Raina in crucial overs (e.g., 10-15 or death overs), adjusting field settings (e.g., close-in fielders for spin) to maximize dismissal chances, especially in playoff scenarios.

### Now Team runs in PowerPlay vs Death Overs

```python

power_play = (
    df[df['over'].between(1, 6)]
    .groupby('batting_team')['batsman_runs']
    .sum()
    .reset_index(name='Powerplay Runs')
)

death_over = (
    df[df['over'].between(16, 20)]
    .groupby('batting_team')['batsman_runs']
    .sum()
    .reset_index(name='Death-Over Runs')
)

team_runs = power_play.merge(death_over, on='batting_team')


fig = px.bar(
    team_runs,
    x='batting_team',
    y=['Powerplay Runs', 'Death-Over Runs'],
    barmode='group',
    title="Team Runs in PowerPlay vs Death-Overs",
    labels={"value": "Runs", "batting_team": "Team", "variable": "Moment"}
)

fig.show()
```

## *Insight*: 
- The  bar plot indicates Mumbai Indians and Chennai Super Kings excel in death overs (8k-10k runs), while teams like Gujarat Lions and Rising Pune Supergiants lag (2k-4k runs). This suggests a gap in finishing ability.  
## *Action*: 
- Struggling teams should recruit or develop death-over specialists (e.g., hard-hitting batsmen like MS Dhoni) and train current players to accelerate in overs 16-20. Leading teams should maintain aggressive intent with their finishers to sustain their edge.
